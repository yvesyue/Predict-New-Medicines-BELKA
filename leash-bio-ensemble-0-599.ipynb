```python
import pandas as pd

sub0435 = pd.read_csv('/kaggle/input/submission-7/submission-7.csv')
sub0432 = pd.read_csv('/kaggle/input/submissions/submission-3.csv')
sub0417 = pd.read_csv('/kaggle/input/submissions/submission-4.csv')
sub09 = pd.read_csv('/kaggle/input/submissions9-10/submission-9.csv')
sub10 = pd.read_csv('/kaggle/input/submissions9-10/submission.csv')

sample_sub = pd.read_csv('/kaggle/input/leash-BELKA/sample_submission.csv')
```

```python
weights = {
    'sub0435': 0.4,
    'sub0432': 0.05,
    'sub0417': 0.05,
    'sub09': 0.25,
    'sub10': 0.25,
}

# Merge the submissions using weighted average
submission_blend = (
    sub0435['binds'] * weights['sub0435'] +
    sub0432['binds'] * weights['sub0432'] +
    sub0417['binds']  * weights['sub0417'] +
    sub09['binds']  * weights['sub09'] +
    sub10['binds'] * weights['sub10']
)

sample_sub['binds']=submission_blend
sample_sub.to_csv('submission.csv', index=False)
sample_sub
```
